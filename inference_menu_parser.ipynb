```python
import requests
import layoutparser as lp
import json
```

```python
from PIL import Image
```

```python
test_json_path = 'assest/outter_split/test.json'
```

```python
f = open(test_json_path)
coco_data = json.load(f)
coco_images = coco_data['images']
```

```python
images_path = [i['file_name'] for i in coco_images]
```

```python
model = lp.Detectron2LayoutModel(
    config_path = "trained-model/fast_outter_region_01/config.yaml",
    model_path = "trained-model/fast_outter_region_01/model_0039999.pth",
    extra_config = ["MODEL.ROI_HEADS.SCORE_THRESH_TEST", 0.8] # <-- Only output high accuracy preds
)
```

```python
test_set_path = "assest/outter_split/" + images_path[5]
```

```python
im = Image.open(test_set_path)
```

```python
layout = model.detect(im)
```

```python
lp.draw_box(im, layout)
```
